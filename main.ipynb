```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

font = {'family' : 'sans-serif',
        'size'   : 15}
matplotlib.rc('font', **font)
```

```python
# define waittime parameters
waitPerEDPatient = 15.0/60.0 # 15 minutes (this is the penalty each patient in traige gets assigned 
                             # if they do not get seen in ED. this was tuned to mimic real-life ED waittimes)
waitPerAmbulance = 20.0/60.0 # 20 minutes (assume MAX ambulance rid is 20 minutes)
                             # this value is scaled by distance to hospital
```

```python
class Patient:
    def __init__(self,location,health):
        self.health = health
        self.location = location
        self.status = 0 # not currently used but can evenutally inform EMT of their suspected COVID patient status
        self.hospital = None
        self.waitTime = 0
        self.waitForED = 0 
        self.waitForICU = 0 
    def updateLocation(self,hospital):
        self.hospital = hospital
    def updateWaitTime(self,time):
        self.waitTime += time
    def setEDTime(self):
        self.waitForED = self.waitTime
    def setICUTime(self):
        self.waitForICU = self.waitTime

class ICU:
    def __init__(self,name,metadata): # passed ICU metadata
        self.name = name
        # currently assuming 1-1 ratio of ICU beds to ICU ventilators for simplicity
        self.max_beds = metadata['max_beds']
        self.max_ventilators = metadata['max_ventilators']
        self.used_beds = 0
        self.used_ventilators = 0
        self.patients = [] # this list keeps track of all ICU patients in hospital
    def updateUsed(self,delta_beds=1, delta_ventilators=1,patient=None):
        self.used_beds += delta_beds
        self.used_ventilators += delta_ventilators
        if (patient != None and delta_beds > 0):
            self.patients.append(patient)
            patient.setICUTime()
        elif (patient != None and delta_beds < 0):
            self.patients.remove(patient)
            
class ED:
    def __init__(self,name,metadata): # passed ER metadata
        self.name = name
        self.max_beds = metadata['max_beds']
        self.used_beds = 0
        self.patients = [] # this list keeps track of all ED patients in hospital
    def updateUsed(self,delta_beds=1,patient=None):
        self.used_beds += delta_beds
        if (patient != None and delta_beds > 0):
            self.patients.append(patient)
            patient.setEDTime()
        elif (patient != None and delta_beds < 0):
            self.patients.remove(patient)
        
class Hospital:
    def __init__(self,name,metadata): # passed hospital metadata
        self.name = name 
        self.location = metadata['location']
        # currenly, hospitals just have ICU and ED but this can easily be added to
        self.ICU = ICU(name=self.name,metadata=metadata['ICU'])
        self.ED = ED(name=self.name,metadata=metadata['ED'])
        self.patients = [] # this list keeps track of all triage (unseen/unadmitted) patients in hospital
    def distanceToPatient(self,location):
        return np.linalg.norm(self.location-location)
    def availableBedED(self):
        return self.ED.max_beds-self.ED.used_beds
    def availableBedICU(self):
        return self.ICU.max_beds-self.ICU.used_beds
    def availableVentilatorICU(self): # this is redudant for now since ICU bed = ICU ventilator
        return self.ICU.max_ventilators-self.ICU.used_ventilators
    def admitToED(self,delta_beds=1,patient=None):
        self.ED.updateUsed(delta_beds=delta_beds,patient=patient)
        if (patient != None):
            self.patients.remove(patient)
    def admitToICU(self,delta_beds=1,delta_ventilators=1,patient=None):
        self.ED.updateUsed(delta_beds=-delta_beds,patient=patient)
        self.ICU.updateUsed(delta_beds=delta_beds,delta_ventilators=delta_ventilators,patient=patient)
    def dischargeFromED(self,delta_beds=1,patient=None):
        self.ED.updateUsed(delta_beds=-delta_beds,patient=patient)
    def dischargeFromICU(self,delta_beds=1,delta_ventilators=1,patient=None):
        self.ICU.updateUsed(delta_beds=-delta_beds,delta_ventilators=-delta_ventilators,patient=patient)
    # this function triages patients and sends them to ED if space and then
    # admits them to ICU if critically-ill (health<0.25)
    def scanAdmit(self):
        # sort patients from sickest to healthiest for triage
        args = np.argsort([p.health for p in self.patients])
        ordered_patients = [self.patients[i] for i in args]
        for i,p in enumerate(ordered_patients):
            if self.availableBedED()>0:
                self.admitToED(patient=p)
                # only 5% of covid patients need ventilation support and 20% need hospitalization.
                # thus 5/20 is 0.25 patients who need to go to the ICU.
                if (p.health<=0.25 and self.availableBedICU()>0):
                    self.admitToICU(patient=p)
            else: break
    # this function is the time-stepper that allows for triage AND probabilistic dischages from hospital
    def updateHospital(self,patient=None,alpha=80.0):
        if (patient != None):
            self.patients.append(patient)
            patient.updateLocation(self)
            # penalize waittime by how far patient had to travel to hospital
            patient.updateWaitTime(waitPerAmbulance*self.distanceToPatient(patient.location))
        # triage
        self.scanAdmit()
        # penalize wait times for patients remaining in triage
        for tempPatient in self.patients:
            tempPatient.updateWaitTime(waitPerEDPatient)
        # discharge patients in ED (only if they are not critically ill), currently there's a [0.31% - 1.25%] of 
        # dischaging patients of health between [0.25 - 1]. since this is updated once at each "timestep" and each
        # "timestep" is 1 patient which we appoximate as 3 min (i.e. a new patient every 3 mins). this corresponds
        # to about a 33% chance of disharging the HEALTHIEST (i.e. health=1) patients within 75 mins. for the 
        # most critically-ill patients eligible for probabilistic discharge (health=0.25), there's only a 7% 
        # chance. intermediate health pateints are interpolated between these values. this is all using the 
        # default alpha value of 80 (see arguments of updateHospital).
        for p in self.ED.patients:
            if p.health>alpha*np.random.random() and p.health>0.25: self.dischargeFromED(patient=p)
        # similar above logic is used for ED but the probabilities are reduced by half such that the MAX chance of
        # leaving the ICU within 75 minutes for the HEALTHIEST patients (health=0.25) is 3%. scales lower with 
        # health. this is the most uncertain parameter since i do not want to explicitly model for recovery/death
        # since it is sad and feels too real right now, and also we dont really have data for this anyway. so ICU
        # "discharge" is a combination of recovery/death that allows for more patients to come into the ICU
        for p in self.ICU.patients:
            if p.health>alpha*np.random.random()/2: self.dischargeFromICU(patient=p)

class City:
    def __init__(self,metadata): # all metadata
        self.hospitals = [] # list of all the hospitals in the city
        for name in metadata.keys():
            self.hospitals.append(Hospital(name=name,metadata=metadata[name]))
    # naive implemtnation (i.e. standard care model of EMT where you just go to the nearest hospital)
    def new_patient_naive(self,patient):
        distances = []
        for hospital in self.hospitals:
            distances += [hospital.distanceToPatient(patient.location)]
        # assign patient to nearest hospital
        h = self.hospitals[np.argsort(distances)[0]]
        h.updateHospital(patient=patient)
        # update all other hospitals so thatthey may perform triage/discharge existing patients
        for hospital in self.hospitals:
            if h != hospital:
                hospital.updateHospital(patient=None)
    # communication implementation (i.e. Ambumetrix with capacity-informed decision making)
    def new_patient(self,patient,alpha=1.0,beta=2.0):
        metric = []
        distances = []
        for hospital in self.hospitals:
            distances += [hospital.distanceToPatient(patient.location)] # IS NOT included in metric currently
            # define metric (ICU is currently weighted to be twice as important as ED bed availabilty
            # scaled by patient's health of course; these parameters will be tuned in real-time to maximize
            # patient treatment and minimize waitime across the city)
            metric += [(alpha*(hospital.availableBedED()))+
                       (beta*((1-patient.health)*(hospital.availableVentilatorICU())))] 
        # assign patient to best hospital according to metric
        h = self.hospitals[np.argsort(metric)[-1]]
        h.updateHospital(patient=patient)
        # update all other hospitals so thatthey may perform triage/discharge existing patients
        for hospital in self.hospitals:
            if h != hospital:
                hospital.updateHospital(patient=None)
    # find patients remaining in traige of each hospital at each "timepoint"
    def get_untreated(self):
        sum_ = 0
        for h in self.hospitals:
            sum_ += len(h.patients)
        return sum_
```

```python
# create patient function with random location and health metric
def createPatient():
    location = np.asarray([np.random.random(),np.random.random()])
    health = np.random.randint(5,995)/1000
    patient = Patient(location=location,health=health)
    return patient

# cross-model comparision via simulation for standard (distance-minimization) and capacity-informed 
# (Ambumetrix, capacity-maximization) models
def simulate(city1,city2,nH,nP,nT=1): 
    ed = np.zeros(nP*nH).reshape([nH,nP]) # ED bed availability over "time"
    wt = np.zeros(nP) # wait time for individual patients for ED
    vt = np.zeros(nP) # wait time for individual patients for ICU
    ht = np.zeros(nP) # health of each patient 
    ut = np.zeros(nP) # untreated patients (remaining in triage)
    ed2 = np.zeros(nP*nH).reshape([nH,nP])
    wt2 = np.zeros(nP) 
    vt2 = np.zeros(nP) 
    ht2 = np.zeros(nP) 
    ut2 = np.zeros(nP)
    for trial in range(nT):
        inpatient1 = []
        inpatient2 = []
        for i in range(nP): 
            patient = createPatient()
            inpatient1.append(patient)
            city1.new_patient(patient)
            for hospital in city1.hospitals:
                j = hospital.name[0]*int(np.sqrt(nH))+hospital.name[1]
                ed[j,i] += hospital.availableBedED()/nT
            ht[i] += patient.health/nT
            ut[i] += city1.get_untreated()/nT
            # second model below
            patient = createPatient()
            inpatient2.append(patient)
            city2.new_patient_naive(patient)
            for hospital in city2.hospitals:
                j = hospital.name[0]*int(np.sqrt(nH))+hospital.name[1]
                ed2[j,i] += hospital.availableBedED()/nT
            ht2[i] += patient.health/nT
            ut2[i] += city2.get_untreated()/nT
        for i,patient in enumerate(inpatient1):
            wt[i] += patient.waitForED/nT
            vt[i] += patient.waitForICU/nT
        for i,patient in enumerate(inpatient2):
            wt2[i] += patient.waitForED/nT
            vt2[i] += patient.waitForICU/nT
    return ed,wt,vt,ht,ut,ed2,wt2,vt2,ht2,ut2
```

```python
fig = plt.figure(figsize=(10,40))

nH = 9 # hospitals
nP = 4000 # patients

metadata = {}
# create hospitals
for i in range(int(np.sqrt(nH))):
    for j in range(int(np.sqrt(nH))):
        name = (i,j)
        maxBed = np.random.randint(20,60)
        metadata[name] = {'location': (np.linspace(0,1,int(np.sqrt(nH)))[i],
                                       np.linspace(0,1,int(np.sqrt(nH)))[j]), 
                          'ED': {'max_beds': maxBed}, 
                          'ICU': {'max_beds':int(0.30*maxBed), 
                          'max_ventilators': int(0.30*maxBed)} }
# currently we are sampling from uniform distributions for random capacity allocation to hospitals, but in big
# cities, this is acutally exponentially distributed. so future analysis with real-world data should be done (!!)
# to get better estimates and the ability to treat patients and their waittimes

city1 = City(metadata=metadata) # our model
city2 = City(metadata=metadata) # naive model
ed1,wt1,vt1,ht1,ut1,ed2,wt2,vt2,ht2,ut2 = simulate(city1,city2,nH,nP,nT=1)

ax = fig.add_subplot(511)
for i in range(nH):
    ax.plot(ed1[i],label='h'+str(i+1))
plt.legend(loc=1)
ax.set_ylabel('average open ED beds')
ax.set_xlabel('cumulative new patients')
ax.set_title('Capacity-Informed Model')
ax.set_ylim([np.min([ed1,ed2]),np.max([ed1,ed2])])

ax = fig.add_subplot(512)
for i in range(nH):
    ax.plot(ed2[i],label='h'+str(i+1))
plt.legend(loc=1)
ax.set_ylabel('average open ED beds')
ax.set_xlabel('cumulative new patients')
ax.set_title('Standard of Care Model')
ax.set_ylim([np.min([ed1,ed2]),np.max([ed1,ed2])])

ax = fig.add_subplot(513)
relInds = wt2 >=0
ax.plot(wt2[relInds],'o',label='SCM')
relInds = wt1 >=0
ax.plot(wt1[relInds],'o',label='CIM')
#ax.plot(np.cumsum(wt1[relInds])/np.linspace(1,len(relInds),len(relInds)))
ax.set_ylabel('ED wait-time')
ax.set_xlabel('cumulative new patients')
ax.set_ylim([np.min([wt1,wt2]),np.max([wt1,wt2])])
plt.legend(loc=1)

ax = fig.add_subplot(514)
relInds = vt2>=0 
ax.plot(ht2[relInds],vt2[relInds],'o',label='SCM')
relInds = vt1 >=0
ax.plot(ht1[relInds],vt1[relInds],'o',label='CIM')
#ax.plot(np.cumsum(vt1)/np.linspace(1,nP,nP))
ax.set_ylabel('ventilator wait-time')
ax.set_xlabel('patient health')
ax.set_ylim([np.min([vt1,vt2]),np.max([vt1,vt2])])
plt.legend(loc=1)

ax = fig.add_subplot(515)
ax.plot(ut2,label='SCM',linewidth=4)
ax.plot(ut1,label='CIM',linewidth=4)
ax.set_ylabel('untreated patients')
ax.set_xlabel('cumulative new patients')
ax.set_ylim([np.min([ut1,ut2]),np.max([ut1,ut2])])
plt.legend(loc=1)


plt.show()
```
